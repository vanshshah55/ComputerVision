```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import random
image =cv2.imread('/content/Baboon.png')
image=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
plt.imshow(image)
```

```python
noise=np.random.normal(50,20,image.shape).astype(np.uint8)
noisy_image=cv2.add(image,noise)
plt.imshow(noisy_image)
```

```python
filter=np.ones((3,3),np.float32)/9
newing=cv2.filter2D(src=noisy_image,ddepth=-1,kernel=filter)
plt.imshow(newing)
```

```python
image=cv2.imread('/content/cameraman.tif')
for a in range(random.randint(3000,10000)):
  i=random.randint(0,len(image)-1)
  j=random.randint(0,len(image)-1)
  clr=random.choice([0,255])
  image[i][j]=clr

greyimage=plt.imshow(image,cmap='grey')
```

```python
denoised=image.copy()
rows, cols = image.shape[:2]
for i in range(rows-3):
  for j in range(cols-3):
    element=image[i:i+3,j:j+3]
    element=element.reshape(-1)
    element.sort()
    denoised[i][j]=element[4]
```

```python
plt.imshow(denoised,cmap='grey')
```

```python
denoiseimg=cv2.fastNlMeansDenoising(src=denoised,h=50)
plt.imshow(denoiseimg)
```
