```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

img = cv2.imread('/content/football.jpg')

b, g, r = cv2.split(img)

plt.figure(figsize=(10, 10))

plt.subplot(1, 3, 1)
plt.imshow(b, cmap='gray')
plt.title('Blue Channel')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(g, cmap='gray')
plt.title('Green Channel')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(r, cmap='gray')
plt.title('Red Channel')
plt.axis('off')

plt.show()
```

```python
dest_not=cv2.bitwise_not(img1)
cv2_imshow(dest_not)
```

```python

img2_resized = cv2.resize(img2, (img1.shape[1], img1.shape[0]))


# Addition
add_img = cv2.add(img1, img2_resized)
cv2_imshow(add_img)

# Subtraction
sub_img = cv2.subtract(img1, img2_resized)
cv2_imshow(sub_img)

# Multiplication
mul_img = cv2.multiply(img1, img2_resized)
cv2_imshow(mul_img)

# Division (with clipping to avoid division by zero)
div_img = cv2.divide(img1, img2_resized)
cv2_imshow(div_img)
```

```python

# Bitwise AND
bitwise_and = cv2.bitwise_and(img1, img2_resized)
cv2_imshow(bitwise_and)

# Bitwise OR
bitwise_or = cv2.bitwise_or(img1, img2_resized)
cv2_imshow(bitwise_or)

# Bitwise XOR
bitwise_xor = cv2.bitwise_xor(img1, img2_resized)
cv2_imshow(bitwise_xor)

# Bitwise NOT (on the first image)
bitwise_not = cv2.bitwise_not(img1)
cv2_imshow(bitwise_not)
```
